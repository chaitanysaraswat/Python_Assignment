# Generate Primes
Q: Generate primes in a range.

Logic: Loop with primality check.

INPUT: Range 1–20

OUTPUT: [2,3,5,7,11,13,17,19]

```python
list1=[]
count=0
for i in range(1,20):
  count=0
  for j in range(1,i+1):
    if i%j==0:
      count+=1
  if count==2:
    list1.append(i)
print(list1)

```

# 11 Maximum Subarray Sum

Q: Max sum of any subarray.

Logic: Kadane's algorithm.

INPUT:[-2,1,-3,4,-1,2,1,-5,4]**bold text**
OUTPUT: 6



```python
list1=[-2,1,-3,4,-1,2,1,-5,4]
max_sum,current_sum=0,0
for i in range (0,len(list1)):
  current_sum+=list1[i]
  if max_sum<current_sum:
    max_sum=current_sum
  if current_sum<0:
    current_sum=0
print(max_sum)
```

# 12 Find All Duplicates
Q: 12 Elements appearing more than once.

Logic: Dictionary for counts.

INPUT:
[4,3,2,7,8,2,1,5,5]

OUTPUT: [2,5]

```python
list1=[4,3,2,7,8,2,1,5,5]
dict1={}
list2=[]
for i in range(0,len(list1)):
    if list1[i] not in dict1:
      dict1[list1[i]]=1
    else:
      dict1[list1[i]]+=dict1[list1[i]]+1
for key,values in dict1.items():
  if values>1:
      list2.append(key)
print(list2)


```
